```python
#here is the key import! 
from IPython.parallel import Client
# turn on plotting
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import num2date
import matplotlib.dates as mdates
#utilities for interacting with the OS
import os

import pyart
def retrieve_qvp(filename, flds = None):
    try:
        if flds == None:
            flds = ['differential_phase', 
                    'cross_correlation_ratio', 
                    'spectrum_width', 
                    'reflectivity', 
                    'differential_reflectivity', 'velocity']
        radar = pyart.io.read(filename)
        desired_angle = 20.0
        index = abs(radar.fixed_angle['data'] - desired_angle).argmin()
        qvp = {}
        for fld in flds:
            this_fld = radar.get_field(index, fld)[:, 0:900].mean(axis = 0)
            qvp.update({fld:this_fld})
        qvp.update({'range': radar.range['data'][0:900], 'time': radar.time})
        x,y,z = pyart.io.common.radar_coords_to_cart(qvp['range']/1000.0, 0.0, 
                                                 radar.fixed_angle['data'][index])
        qvp.update({'height': z})
        del radar
    except:
        qvp = 'NO'
    return qvp
```

```python
indir = '/data-in2/sc8/lassen_CPOL/cpol_0506/'
all_csapr_files = os.listdir(indir)
all_csapr_files.sort()
interesting_files = []
for fn in all_csapr_files:
    if 'PPI' in fn:
        if os.stat(indir + fn).st_size > 5e6:
            if fn[0] != '.':
                interesting_files.append(indir + fn)
interesting_files.sort()
```

```python
My_Cluster = Client()
My_View = My_Cluster[:]
print My_View
print len(My_View)
```

```python
#Turn off blocking so all engines can work async
My_View.block = False

#on all engines do an import of Py-ART
My_View.execute('import pyart')

#Map the code and input to all workers
result = My_View.map_async(retrieve_qvp,interesting_files[3000:8000])

#Reduce the result to get a list of output
qvps = result.get()
```

```python
qvps2 = []
for qv in qvps:
    if qv != 'NO':
        qvps2.append(qv)

#create an empty dictionary to be filled
time_series = {}

#work out how many profiles
n_times = len(qvps2)

#Loop over all fields 
for fld in ['differential_phase', 
            'cross_correlation_ratio',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']:
    #Create empty time series
    this_fld = np.zeros([n_times, len(qvps2[0][fld])])
    
    #create time array
    times = np.zeros(n_times)
    
    #Loop over all times
    for i in range(n_times):
        if qvps2[i] != 'NO':

            #Create a datetime object
            dateobj = num2date(qvps2[i]['time']['data'][0], qvps2[i]['time']['units'])

            #Append a numerical date using Matplotlib's time reference
            times[i] = mdates.date2num(dateobj)

            #Update the profile to the time series
            this_fld[i,:] = qvps2[i][fld]

    #append the time series to the dictionary
    time_series.update({fld: this_fld[times.argsort(), :]})
times.sort()
```

```python
fig = plt.figure(figsize = [15,5])
pc = plt.pcolormesh(times, qvps[0]['height']/1000.0, 
                    time_series['reflectivity'].transpose(),
                    vmin = -20, vmax = 64)
ax = plt.gca()
ax.xaxis_date()
#Magically deal with time!
fig.autofmt_xdate()
plt.xlabel('Time (UTC)')
plt.ylabel('Height (km)')
plt.ylim([0,20])
cb = plt.colorbar()
cb.set_label('Mean Reflectivity factor (dBZ)')
```
