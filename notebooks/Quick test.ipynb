```python
#here is the key import! 
from IPython.parallel import Client
from IPython.display import display, clear_output
import time
# turn on plotting
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from netCDF4 import num2date
import matplotlib.dates as mdates
#utilities for interacting with the OS
import os

import pyart
def retrieve_qvp(filename, flds = None):
    if flds == None:
        flds = ['differential_phase', 
            'cross_correlation_ratio',
            'spectrum_width',
            'reflectivity', 'differential_reflectivity']
    radar = pyart.io.read(filename)
    desired_angle = 20.0
    index = abs(radar.fixed_angle['data'] - desired_angle).argmin()
    qvp = {}
    for fld in flds:
        this_fld = radar.get_field(index, fld)[:, 0:900].mean(axis = 0)
        qvp.update({fld:this_fld})
    qvp.update({'range': radar.range['data'][0:900], 'time': radar.time})
    x,y,z = pyart.io.common.radar_coords_to_cart(qvp['range']/1000.0, 0.0, 
                                             radar.fixed_angle['data'][index])
    qvp.update({'height': z})
    del radar
    return qvp
```

```python
indir = '/data-in2/sc8/lassen_CPOL/cpol_0506/'
```

```python
all_csapr_files = os.listdir(indir)
all_csapr_files.sort()
interesting_files = []
for fn in all_csapr_files:
    if 'PPI' in fn:
        if os.stat(indir + fn).st_size > 5e6:
            if fn[0] != '.':
                interesting_files.append(indir + fn)
interesting_files.sort()
```

```python
radar = pyart.io.read(interesting_files[10032])
```

```python
gatefilter = pyart.correct.GateFilter(radar)
gatefilter.exclude_below('reflectivity', -5)
```

```python
corr_vel = pyart.correct.dealias_region_based(
    radar, vel_field='velocity', keep_original=False, 
skip_along_ray=100, skip_between_rays=100, gatefilter = gatefilter)
```

```python
radar.add_field('corrected_mean_doppler_velocity', corr_vel, True)
```

```python
my_display = pyart.graph.RadarDisplay(radar)
```

```python
#subplot(nrows, ncols, plot_number)
fig = plt.figure(figsize = [12,5])
sb1 = plt.subplot(121)
my_display.plot_ppi('corrected_mean_doppler_velocity', 0,
                 vmin = -20, vmax =20)
sb1 = plt.subplot(122)
my_display.plot_ppi('velocity', 0,
                 vmin = -20, vmax =20)
```
